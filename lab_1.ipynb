```python
nums = [2, 7, 11, 15]
target = 9


def two_sum(nums, target):
    """
    Находит два индекса чисел в списке, которые дают в сумме target

    Аргументы:
      nums: список целых чисел
      target: целое число

    Возвращает:
      список из двух индексов, если подходящая пара найдена
      None/2, если пара не найдена/допущена ошибка при вводе данных
    """
    # Проверка входных данных
    if len(nums) < 2:
        return 2

    # Проверка, что все элементы - целые числа
    for num in nums:
        if not isinstance(num, int):
            return 2

    # Проверка, что target - целое число
    if not isinstance(target, int):
        return 2

    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            if nums[i] + nums[j] == target:
                return [i, j]



if two_sum(nums, target) == 2:
    print('Ощибка. Перечитайте условие задачи')
elif two_sum(nums, target) == None:
    print('Pешений нет')
else:
    print(two_sum(nums, target))

assert (two_sum(nums, target)) == [0, 1], "Провалено"
```
